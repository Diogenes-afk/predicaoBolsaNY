### começando preparação para treino de modelo

#### separando ultima linha para teste

```python
amanha = df[-1::]
```

```python
df = df.drop(df[-1::].index, axis=0)
```

```python
baseF = df
```

```python
baseF.head()
```

#### Criando coluna target para que recebe o valor de fechamento do dia seguinte

```python
baseF['Target'] = baseF['Close'][1:len(baseF)].reset_index(drop=True)
```

```python
baseF[-2::]
```

#### sepanrando valor para treino

```python
prev = baseF[-1::].drop('Target', axis = 1)
prev
```

```python
treino = baseF.drop(baseF[-1::].index, axis = 0)
treino.tail()
```

#### Se o valor target for maior que o da coluna Close do dia passado, altera para 1, simulando um boolean que corresponder a ação subir

```python
treino.loc[treino['Target'] > treino['Close'], 'Target'] = 1
treino.tail()
```

```python
treino.loc[treino['Target'] != 1, 'Target'] = 0
treino.tail()
```

```python
treino.tail()
```

#### data_sample = treino.sample(n=1000, random_state=1)

### Importando modelo de MachineLearning

```python
y = treino['Target']
x = treino.drop('Target', axis= 1)

from sklearn.model_selection import train_test_split
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size = 0.3)

from sklearn.ensemble import ExtraTreesClassifier
modelo = ExtraTreesClassifier()
modelo.fit(x_treino, y_treino)

resultado = modelo.score(x_teste, y_teste)
print("Resultado = ", resultado)
```
